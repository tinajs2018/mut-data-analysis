```python
#matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
df.head(5)
```

```python
plt.xlabel("cat1marks")
plt.ylabel("cat2marks")
plt.title("The results of cat1 againist cat2")
y=df['cat1']
x=df['cat2']
plt.plot(x,y ,color="green",linewidth=5,linestyle="dotted")
```

```python
y=df ['main']
x=df ['cat1']
plt.plot(y,x)
```

```python
plt.xlabel('marks of the cat one and cat2')
plt.ylabel("the total marks of the whole exams")
plt.title("A graph of  total marks against cats marks")
plt.plot(df['total'],df['cat1'], label='cat1' )
plt.plot(df['total'],df['cat2'], label= 'cat2' )
plt.plot(df['total'],df['cat3'] ,label= 'main exam' )
plt.legend(loc='upper right' ,shadow =True ,fontsize='large') 
plt.grid()
```

```python
import numpy as np
xpos=np.arange(len(df['course']))
xpos
```

```python
plt.ylabel("The marks for main exam(100%)")
plt.title("The marks across all schools")
plt.xticks(xpos, df['course'])
plt.bar(xpos +0.2,df['main'] ,width=0.4 ,label="main")
plt.bar(xpos -0.2,df['cat1'] ,width=0.4 ,label="cat1")
plt.grid()
plt.legend()
```

```python
#drawing the bar cahart into a horizontal position
plt.title("the result of mut students")

plt.yticks(xpos, df['course'])
plt.barh(xpos +0.2,df['main'],label="main ")
plt.barh(xpos -0.2,df['cat1'] ,label="cat1")

plt.legend()
```

```python
#dealing with histograms
plt.title('Histogram of cat 1 and cat2')
plt.hist([df['cat3'],df['cat2'] ],bins=[0,40,60,89],rwidth=0.95 ,color=['g', 'blue'], label=['cat1','cat2'], orientation='horizontal')
plt.legend()
```

```python
#dealing with pie
plt.axis("equal")
plt.pie(df['total'] ,labels=df['course'] ,radius=2, autopct= '% 0.1f%%',explode=[0.5,0,0,0,0.5,0,0,0,0,0])
plt.show()
```
